# Training of the Dataset

```python
X_train = df.drop('Species', axis=1) # Change 'target' to 'Species'
y_train = df.iloc[:, -1]

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python

from sklearn.tree import plot_tree
plot_tree(dt)
```

# Classification

```python
import dtreeviz
from sklearn.tree import DecisionTreeClassifier


from sklearn.preprocessing import LabelEncoder   # LABEL ENCODING
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)

dt.fit(X_train, y_train_encoded)

# Generate the visualization:
viz = dtreeviz.model(dt,
                    X_train,
                    y_train_encoded,  # Use encoded labels
                    feature_names=X_train.columns,
                    target_name='Species',
                    class_names=le.classes_,  # Use original class names
                    )

viz.view(scale=1.5)  # Display the visualization
```

Great thing is We can see the Splitting Criteria on the graph like first split on the basis of Petal width less the 0.80 then the split on the basis of Petal width 1.75
